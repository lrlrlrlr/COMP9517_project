```python
import pandas
from ultilities import *
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

count_toba = len(toba_filenames)
count_arab = len(arab_filenames)
ax = sns.barplot(x=['Tobacco','Arabidopsis'], y=[count_toba,count_arab])
ax.set_title("Sample Counts",size=20)

```

```python
# analysis picture size

arab_imgs = []
toba_imgs = []
for filename in arab_filenames:
    img = cv.imread(path_arab + '\\' + filename, 0)
    arab_imgs.append(img.shape)

for filename in toba_filenames:
    img = cv.imread(path_toba + '\\' + filename, 0)
    toba_imgs.append(img.shape)
```

```python
df = pandas.DataFrame(arab_imgs)
df.columns=['x','y']
df['size'] = df.x*df.y
df['type'] = 'Arabidopsis'

df2 = pandas.DataFrame(toba_imgs)
df2.columns=['x','y']
df2['size'] = df2.x*df2.y
df2['type'] = 'Tobacco'

```

```python
df3 = df.append(df2)
df3
```
